```python
import csv
import requests
from bs4 import BeautifulSoup 
```

```python
data = [] #[ [0  website, 1 Name, 2 Model, 3 Stroke, 4 Voltage, 5 Dynamic, 6 Static, 7 Speed (Full), 8 Feedback, 9 weather Protection, 10 Temp], [] ]
```

```python
site = "https://www.progressiveautomations.ca/"
page = "https://www.progressiveautomations.ca/collections/linear-actuators"

html = requests.get(page)
if html.status_code==200:
    soup = BeautifulSoup(html.text, 'lxml')
    results = [a['href'] for a in soup.find_all('a',class_='catalog-page__item-link', href=True) if a.text]
```

```python
soup
```

```python
for result in results:
    info = []
    item = requests.get(site+result)
    if html.status_code==200:
        soup = BeautifulSoup(item.text, 'lxml')
        specs = [spec.text for spec in soup.find_all('div', class_="catalog-info__table-row-td")]
        
        info.append(site+result)
        info.append(soup.find(class_='catalog-item__info-header').text)
        info.append(soup.find(class_='catalog-item__info-model').text) 
        
        for x in range(len(specs)):
            if specs[x] == 'Input Voltage:':
                info.append(specs[x+1])
            if specs[x] == 'Stroke:':
                info.append(specs[x+1])
            if specs[x] == 'Feedback:':
                info.append(specs[x+1])
            if specs[x] == 'Weather Protection:':
                info.append(specs[x+1])
            if specs[x] == 'Operational Temperature:':
                info.append(specs[x+1])
        data.append(info)
```

```python
with open("progressiveAutomations.csv", "w+") as fileWriter:
    wr = csv.writer(fileWriter)
    wr.writerows(data)
```
