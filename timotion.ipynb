```python
import csv
import requests
from bs4 import BeautifulSoup
```

```python
site = "https://www.timotion.com"
page = "https://www.timotion.com/en/products/intro/linear-actuators/lists?guid=1481269298"

html = requests.get(page)
if html.status_code==200:
    soup = BeautifulSoup(html.text, 'lxml')
    div = [div for div in soup.find_all('div',class_='product-text', href=True) if a.text]
    links = [a['href'] for a in soup.find_all('a', href=True) if a.text]
    links = links[78:120]
```

```python
#for link in links:
data = []

for link in links:
    html = requests.get(site+link)
    if html.status_code==200:
        soup = BeautifulSoup(html.text, 'lxml')
        title = (soup.find('h1').text)
        specs = [spec.text for spec in (soup.find('ul', class_="dot")).find_all('li')]

        specs.insert(0, title)
        data.append(specs)
```

```python
with open("timotion.csv", "w+") as fileWriter:
    wr = csv.writer(fileWriter)
    wr.writerows(data)
```
